### Correlation

- To see how many features are there which are correlated to each other.
- For checking correlation we use a statistical method that finds the bonding and relationship between two features.

```python
# Heatmap
plt.figure(figsize=[19,10])
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

- We will find those features which are fully correlated to each other.
- By doing this we can reduce the number of features from the data.
- We have to discard those correlated, because relationship among them is equal they equally impact on model accuracy so, we delete one of them.

```python
df.corr(numeric_only=True).columns
```

```python
# Function for finding those featues whose correlation number is high

for a in range(len(df.corr(numeric_only=True).columns)):
    for b in range(a):
        if abs(df.corr(numeric_only=True).iloc[a,b])>0.7:
            names = df.corr(numeric_only=True).columns[a]
            print(names)
```

- total sulfur dioxide satisfies the condition
- So, will drop this feature

```python
df1 = df.drop('total sulfur dioxide', axis=1)
```

```python
df1.head()
```

### Handle null values

```python
df1.isnull().sum()
```

- We see that there are not many null values are present in our data so we simply fill them with the help of the fillna() function.

```python
df1.update(df1.fillna(df1.mean(numeric_only=True)))
```

```python
df1.isnull().sum()
```

```python
# For categorical variables

df1 = pd.get_dummies(df1, drop_first=True)
df1
```
